```python
x1=open("insulin1.fasta","r")
x2=x1.readlines()
seq={}
for j in x2:
    j=j.rstrip("\n")
    xyz=re.match('>',j)
    if(xyz):
        header=j
        seq[j]=''
    else:
        seq[header]=seq[header]+j
print(seq)
```
