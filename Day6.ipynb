```python
abc=open("gene_new.txt","r")
pqr=abc.readlines()
embs={}
for i in pqr:
    xyz=i.split();
    lmn=xyz[0]
    if (lmn not in embs.keys()):
        embs[lmn]=1;
    else:
        embs[lmn]=embs[lmn]+1
print(embs)
```
