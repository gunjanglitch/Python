```python
abc=open("brca1.fasta","r")
pqr=abc.readlines()
seq=[]
for i in pqr:
    if(i[0]=='>'):
        header=i[0:]
    else:
        seq+=i.rstrip()

x=0
A=[]
T=[]
G=[]
C=[]

for j in seq:
    if(j[x]=='A'):
        A.append('A')
    elif(j[x]=='T'):
        T.append('T')
    elif(j[x]=='G'):
        G.append('G')
    elif(j[x]=='C'):
        C.append('C')
    else:
        x+=1

print("Number of A:", len(A))
print("Number of T:", len(T))
print("Number of G:", len(G))
print("Number of C:", len(C))

total=len(A)+len(T)+len(G)+len(C)
print("The length of gene", header, "is", "total", "bp")

gc=len(G)+len(C)
at=len(A)+len(T)
gcp=(gc/total)*100
atp=100-gcp
print("GC percentage is:", round(gcp,2),'%')
print("AT percentage is:", round(atp,2), '%')

mw=(len(G)*329.21)+(len(A)*313.21)+(len(T)*304.2)+(len(C)*289.18)
print("Molecular Weight of gene is:", round(mw,2))
```
